# Task 1

```python
import numpy as np
import matplotlib.pyplot as plt

days = np.arange(1, 31)
prices = np.cumsum(np.random.randn(30) * 2 + 100)


plt.figure(figsize=(10, 5), dpi=100) #figure size and resolution

plt.plot(days, prices, marker = 'o', markersize = 6, linestyle = '-', color = 'b', label = 'Stock Price')
plt.xlabel('Days')
plt.ylabel('Prices (USD)')
plt.title('Stock Market Trend')
plt.legend()
plt.grid( True)

plt.show()
```

# Task 2

```python
days = np.arange(1, 11)
temp_day = np.random.randint(25, 35, size = 10)
temp_night = np.random.randint(15, 25, size = 10)

plt .figure(figsize=(8, 5))

plt.plot(days, temp_day, marker = 'o', markersize = 8, linestyle = '--', color = 'r', label = 'Day Temprature')
plt.plot(days, temp_night, marker = 'o', markersize = 8, linestyle = '-', color = 'b', label = 'Night Temprature')

plt.xlabel('Days')
plt.ylabel('Temprature (Celsius)')
plt.title(' Daily Temperature Variation')
plt.legend()
plt.grid(True)


plt.show()
```

# Task 3

```python
countries = ['Brazil', 'USA', 'China', 'India', 'Germany']
population = np.array([213, 331, 1441, 1380, 83])
pop_growth = np.array([0.9, 0.8, 1.2, 1.5,  0.5])

plt.figure(figsize = (8, 5))

plt.scatter(pop_growth, population, c = population, cmap = 'viridis', s = population*2, alpha = 0.6)

plt.xlabel("Population Growth (%)")
plt.ylabel("Total Population (millions)")
plt.title("Population Growth vs. Total Population")

for i, country in enumerate(countries):
    plt.annotate(country, (pop_growth[i], population[i]), fontsize = 10, ha = 'right')

plt.colorbar(label="Population (millions)")
plt.grid(True)
plt.show()
```

# Task 4

```python
np.random.seed(42)
age = np.random.randint(18, 60, size = 100)
spending_score = np.random.randint(20, 100, 100)
cust_categories = np.random.choice([1, 2, 3, 4], 100)

plt.figure(figsize = (8, 5))

scatter = plt.scatter(age, spending_score, c = cust_categories, cmap = 'jet', alpha = 0.7)

plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation Analysis')

plt.colorbar(scatter, label = 'Customer Category')
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(42)

region = np.random.randint(1, 6, 50) # 5 regions
sales = np.random.randint(500, 10000, 50)
profit = np.random.randint(50, 1000, 50)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(region, sales, profit, c=profit, cmap='coolwarm', s=50, alpha=0.7)

ax.set_xlabel("Region")
ax.set_ylabel("Sales")
ax.set_zlabel("Profit")
ax.set_title("3D Sales Performance Analysis")
fig.colorbar(scatter, label="Profit")

plt.show()
```
